```python
def load_dataset(path):

    df = pd.read_csv(path)
    df = df.fillna("missing")

    possible_labels = [
        "label","Label",
        "attack_cat","Attack_cat",
        "class","Class",
        "target","Target",
        "is_attack","attack"
    ]

    label_col = None
    for c in possible_labels:
        if c in df.columns:
            label_col = c
            break

    if label_col is None:
        raise ValueError("No label column found")

    if label_col.lower() == "attack_cat":
        y = (df[label_col] != "Normal").astype(int).values
    else:
        y = df[label_col].values

    drop_cols = [c for c in ["label", "attack_cat"] if c in df.columns]

    X = df.drop(columns=drop_cols)

    cat_cols = X.select_dtypes(include=["object"]).columns
    for c in cat_cols:
        X[c] = LabelEncoder().fit_transform(X[c].astype(str))

    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    return X, y
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

def load_dataset(path):
    # Standard column names for the UNSW-NB15 dataset
    col_names = ['srcip', 'sport', 'dstip', 'dsport', 'proto', 'state', 'dur', 'sbytes', 'dbytes', 'sttl', 'dttl', 'sloss', 'dloss', 'service', 'sload', 'dload', 'spkts', 'dpkts', 'swin', 'dwin', 'stcpb', 'dtcpb', 'smeansz', 'dmeansz', 'trans_depth', 'res_bdy_len', 'sjit', 'djit', 'sintpkt', 'dintpkt', 'tcprtt', 'synack', 'ackdat', 'is_sm_ips_ports', 'ct_srv_src', 'ct_con_dst', 'ct_src_ltm', 'ct_srv_dst', 'ct_dst_ltm', 'ct_src_dport_ltm', 'ct_dst_sport_ltm', 'ct_dst_src_ltm', 'is_ftp_login', 'ct_ftp_cmd', 'ct_flw_http_mthd', 'ct_src_ltm_double', 'ct_srv_dst_double', 'attack_cat', 'label']

    # Read the CSV with no header and assign the correct column names
    df = pd.read_csv(path, header=None, names=col_names)
    df = df.fillna("missing")

    possible_labels = [
        "label","Label",
        "attack_cat","Attack_cat",
        "class","Class",
        "target","Target",
        "is_attack","attack"
    ]

    label_col = None
    for c in possible_labels:
        if c in df.columns:
            label_col = c
            break

    if label_col is None:
        raise ValueError("No label column found")

    if label_col.lower() == "attack_cat":
        y = (df[label_col] != "Normal").astype(int).values
    else:
        y = df[label_col].values

    X = df.drop(columns=[label_col])

    cat_cols = X.select_dtypes(include=["object"]).columns
    for c in cat_cols:
        X[c] = LabelEncoder().fit_transform(X[c].astype(str))

    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    return X, y

TRAIN_PATH = "/content/UNSW_NB15_training-set.csv"
TEST_PATH  = "/content/UNSW_NB15_testing-set.csv"


X_train, y_train = load_dataset(TRAIN_PATH)
X_test,  y_test  = load_dataset(TEST_PATH)


print("Train:", X_train.shape)
print("Test :", X_test.shape)

print("Train attack ratio:", y_train.mean())
print("Test attack ratio :", y_test.mean())
```

```python
class DefenseEnv(gym.Env):

    def __init__(self, X, y):
        super().__init__()
        self.X = X
        self.y = y
        self.i = 0

        self.observation_space = spaces.Box(
            low=-5, high=5,
            shape=(X.shape[1],),
            dtype=np.float32
        )

        self.action_space = spaces.Discrete(4)

    def reset(self, seed=None, options=None):
        self.i = 0
        return self.X[self.i], {}

    def step(self, action):

        label = self.y[self.i]

        if label == 0:
            reward = 1 if action == 0 else -2
        else:
            reward = [ -5, 1, 3, 5 ][action]

        self.i += 1
        done = self.i >= len(self.X)-1

        return self.X[self.i], reward, done, False, {}
```

```python
env = DefenseEnv(X_train, y_train)

model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=50000)
```

```python
env_test = DefenseEnv(X_test, y_test)

obs, _ = env_test.reset()

preds = []
true = []

done = False

while not done:
    action, _ = model.predict(obs)
    preds.append(action)
    true.append(y_test[env_test.i])

    obs, _, done, _, _ = env_test.step(action)
```
